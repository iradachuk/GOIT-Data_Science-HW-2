```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", 
                   match='Регіон')
tmp = tmp[0]
print(tmp)
```

Вивести перші рядки таблиці за допомогою методу head

```python
print(tmp.head())
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
tmp.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
tmp = tmp.replace('—', np.NaN)
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
tmp.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
tmp['2014'] = pd.to_numeric(tmp['2014'])
tmp['2019'] = pd.to_numeric(tmp['2019'])

tmp.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
tmp.isnull().sum()/tmp.shape[0]
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
tmp = tmp.drop(27)
tmp
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
tmp = tmp.fillna({'1950': tmp['1950'].mean(), '1960': tmp['1960'].mean(), '1970': tmp['1970'].mean(), '1990': tmp['1990'].mean(), '2000': tmp['2000'].mean(),
                   '2012': tmp['2012'].mean(), '2014': tmp['2014'].mean(), '2019': tmp['2019'].mean()})
tmp
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
tmp[tmp['2019'] > tmp['2019'].mean()][['Регіон', '2019']]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
tmp[['Регіон', '2014']].max()
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
tmp['2019'].plot.bar()
plt.show()
```

<b>Візуалізація</b>

```python
fig, ax = plt.subplots(figsize=(10, 7))
date = tmp.loc[tmp['Регіон'] == 'Львівська', '1950':'2019'].columns.to_list()
k_per_year = tmp.loc[tmp['Регіон'] == 'Львівська', '1950':'2019'].values[0]
plt.plot(date, k_per_year, label='Коефіцієнт народжуваності', linestyle='-', marker='*')
plt.xlabel('Рік', fontsize=15)
plt.ylabel('Коефіцієнт народжуваності', fontsize=15)
plt.title('Коефіцієнт народжуваності в Львівській області', fontsize=15)
plt.xlim(0, 7)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

for index, row in tmp.iterrows():
    region = row['Регіон']
    k_per_year = row['1950':'2019'].values
    plt.plot(date, k_per_year, label=region, linestyle='-', marker='*')

plt.xlabel('Рік', fontsize=15)
plt.ylabel('Коефіцієнт народжуваності', fontsize=15)
plt.title('Коефіцієнт народжуваності в областях України', fontsize=15)
plt.xlim(0, 7)
plt.legend(fontsize=6)
plt.show()
```

```python
labels = tmp['Регіон']
sizes = tmp.loc[:, '2019']

fig, ax = plt.subplots(figsize=(10, 10))
ax.pie(sizes, labels=labels, autopct='%.2f%%', startangle=90)
ax.axis('equal')
plt.title("Відсоток народжуванності 1960 року по областях", y=1.08)
plt.subplots_adjust(bottom=0.1)
plt.show()
```
