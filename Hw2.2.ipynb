Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
devs = devs.rename(columns={'Должность': 'position', 'Язык.программирования': 'programming_language', 'Общий.опыт.работы': 'general_experience',
                            'Опыт.работы.на.текущем.месте': 'current_experience', 'Зарплата.в.месяц': 'month_salary', 'Изменение.зарплаты.за.12.месяцев':
                            'salary_changing', 'Город': 'city', 'Размер.компании': 'company_size', 'Тип.компании': 'company_type', 
                            'Пол': 'sex', 'Возраст': 'age', 'Образование': 'education', 'Еще.студент': 'student', 'Уровень.английского': 'english_level',
                            'Дата.заполнения': 'registration_date'})
```

```python
python_data = devs.query("programming_language == 'Python'")
python_data
```

Визначте розмір таблиці python_data за допомогою методу shape

```python
python_data.shape
```

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data = python_data.groupby('position')
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
new_df = python_data['month_salary'].agg(['min', 'max'])
new_df
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(data):
    return (data['min'] + data['max'])/2

avg = new_df.apply(fill_avg_salary, axis=1)
avg.name = 'avg'
new_df = pd.concat([new_df, avg], axis=1)
new_df
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
new_df['avg'].describe()
```

Збережіть отриману таблицю в CSV файл

```python
new_df.to_csv('2017_jun_final_new.csv')
```

<b>Візуалізація</b>

```python
languages = devs['programming_language'].value_counts(normalize=True)

other_languages = languages[languages < 0.02].sum()
languages = languages[languages >= 0.02]
languages['Інші'] = other_languages

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(languages.values, labels=languages.index, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Мови програмування за популярністю')
plt.show()
```

```python
grouped = devs.groupby('programming_language')['month_salary'].mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(grouped['programming_language'], grouped['month_salary'], color='#9966FF')

ax.set_xlabel('Мова програмування', fontsize=15, color='blue')
ax.set_xticklabels(grouped['programming_language'], rotation=45)

ax.set_ylabel('Середня зарплата на місяць', fontsize=15, color='blue')

ax.set_title('Залежність середньої заробітньої платні від мови програмуівання', fontsize=15)
plt.show()
```

```python
bins = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]

plt.hist(devs['month_salary'], bins=bins, edgecolor='black')

plt.xlabel('Зарплата на місяць')
plt.ylabel('Кількість спостережень')

plt.title('Розподіл зарплат')

plt.show()
```
